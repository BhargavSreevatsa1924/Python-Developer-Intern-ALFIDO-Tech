Program to Create a To-Do List 

```python
# Simple To-Do List App using only basic data structures

tasks = []   # list to store all task dictionaries

def show_menu():
    print("\n========= TO-DO LIST MENU =========")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Completed")
    print("4. Delete Task")
    print("5. Exit")
    print("="*40)

def add_task():
    task_name = input("Enter task name: ")
    task = {"task": task_name, "status": "Pending"}
    tasks.append(task)
    print(" Task added!")

def view_tasks():
    if not tasks:
        print("No tasks available.")
        return
    
    print("\n------ Your Tasks ------")
    for i, task in enumerate(tasks, start=1):
        print(f"{i}. {task['task']} - {task['status']}")
    print("------------------------")

def mark_completed():
    view_tasks()
    if not tasks:
        return
    try:
        num = int(input("Enter task number to mark completed: "))
        tasks[num - 1]["status"] = "Completed"
        print(" Task marked as Completed!")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task['task']} - {task['status']}")
    except (ValueError, IndexError):
        print(" Invalid task number")

def delete_task():
    view_tasks()
    if not tasks:
        return
    try:
        num = int(input("Enter task number to delete: "))
        tasks.pop(num - 1)
        print("Task deleted!")
    except (ValueError, IndexError):
        print("Invalid task number")

# Main program loop
while True:
    show_menu()
    choice = input("Enter your choice (1-5): ")

    if choice == "1":
        add_task()
    elif choice == "2":
        view_tasks()
    elif choice == "3":
        mark_completed()
    elif choice == "4":
        delete_task()
    elif choice == "5":
        view_tasks()
        print("bye! See ya later!!..")
        break
    else:
        print("Invalid choice. Try again.")
```
