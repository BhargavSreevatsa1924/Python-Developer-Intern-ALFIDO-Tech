```python
import random
print("welcome to the guessing games!!")
print("you will be allowed to have 3 chances!! to guess between the numbers 1 and 100")

num = random.randint(1,100)
chances = 3
Guess_count = 0

# the game starts
while Guess_count < chances:
     Guess_count += 1
     guess = int(input("enter your guess: "))
     
     
     if guess == num:
         print("Congratulation ,correct number is {0} and your guess is {1} matching ".format(num,guess))
         break
     elif Guess_count >= chances and guess != num:
         print("Sorry your guess was {0} but the Random Number was {1} =[0].Better luck next time".format(guess,num))
         print("you have reached the limit of 3 chances , Try Again Next time!!")
         
  
     elif guess > num :
         print("your guess: {0} was too high.Try a lower number".format(guess))
     elif guess < num :
          print("your guess: {0} was too low.Try a Upper number".format(guess))
   
         
     
         
         
    
    

```
